```python
image = cv2.imread(r"C:\Users\91830\Downloads\Data Science  Course\Machine Learning\Images PRocessing\sample parrot img.jpg")
gray = cv2.cvtColor(image,cv2.COLOR_BGRA2GRAY) # to chage into another color.

cv2.imshow('bird',gray) # for displaying the image
cv2.waitKey(0) #time(ms) to show the image , using 0 will keep the img untile we colse the window.
cv2.destroyAllWindows() #close the window automatically
```

```python
hls = cv2.cvtColor(image,cv2.COLOR_BGR2HLS) # to chage into another color.

cv2.imshow('bird',hls) # for displaying the image
cv2.waitKey(0) #time(ms) to show the image , using 0 will keep the img untile we colse the window.
cv2.destroyAllWindows() #close the window automatically
```

```python
rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB) # to chage into another color.

cv2.imshow('bird',rgb) # for displaying the image
cv2.waitKey(0) #time(ms) to show the image , using 0 will keep the img untile we colse the window.
cv2.destroyAllWindows() #close the window automatically
```

# Creating our own img

```python
import numpy as np
```

```python
black = np.zeros((500,700,3))
cv2.imshow('Black',black)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
white = np.ones((500,700,3))
cv2.imshow('White',white)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Creating symbols

```python
white_rec = np.ones((500,700,3))
cv2.rectangle(img = white_rec, pt1 = (100,100), pt2 = (150,150), color = (0,0,255), thickness = -1)
cv2.circle(img = white_rec, center = (250,350), radius = 50, color = (0,0,255), thickness = -1)
cv2.line(img = white_rec, pt1 = (0,0), pt2 = (1000,700), color = (0,0,255), thickness = 2)
cv2.imshow('White_rec',white_rec)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Creating Text

```python
black = np.zeros((500,700,3))

cv2.putText(img = black, text = "Computer Vision",
           org = (100,100),
           fontFace = cv2.FONT_HERSHEY_PLAIN,
           fontScale = 2,
           color = (255,255,255),
           thickness = 2)

cv2.imshow('Text',black)
cv2.waitKey(0)
cv2.destroyAllWindows()            
```
