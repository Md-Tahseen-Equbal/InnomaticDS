# Density / KDE Plot  (Kernal Density Plot)
-  On Numerical Data 
- **Purpose** - Data Distribution
-  We are representing probability value on y-axis( Ranges from 0 to 1)

```python
df.drop('unnamed1',
  axis='columns', inplace=True)
```

```python
sns.displot(data = df ,x = 'Amount',kind='kde',fill=True,hue='Age Group')
```


1. **General Observations:**
   - The majority of amounts for all age groups are positive, with the highest densities clustering around the lower end of the amount range.
   - There are few instances of negative amounts, which may be considered as outliers or special cases.

2. **Age Group Insights:**
   - **26-35 Age Group:** This group has the highest density peak around the 10,000 mark, indicating that individuals in this age range are most likely to have amounts close to this value.
   - **36-45 Age Group:** This group also shows a high density around the same range as the 26-35 age group but slightly lower, suggesting similar financial behavior or circumstances.
   - **46-50 Age Group:** The density for this group is slightly lower than the 26-35 and 36-45 age groups but follows a similar distribution pattern.
   - **18-25 Age Group:** This group has a lower density overall compared to the older groups, indicating smaller amounts or fewer instances.
   - **51-55 Age Group:** This group has one of the lower density distributions, suggesting smaller amounts or fewer instances.
   - **55+ Age Group:** This group shows the lowest density, indicating even smaller amounts or fewer instances.
   - **0-17 Age Group:** This group has a very low density, which is expected due to lower financial activities typically associated with this age range.

3. **Amount Distribution:**
   - The peak amounts for most age groups are centered around 5,000 to 10,000.
   - There are secondary peaks and fluctuations indicating other common amount ranges, but these are less pronounced.
