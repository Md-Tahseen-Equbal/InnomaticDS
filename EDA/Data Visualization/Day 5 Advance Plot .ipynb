# Advance Plot

```python
import seaborn as sns 
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Violin Plot 
1. THe combination of the density plot and box plot is violin plot
2. What we drive insight from violin plot are **Distributuion and Outliers**

```python
sns.violinplot(data = df, x= 'total_bill')   # Combination of Density and Box Plot
```

```python
sns.violinplot(data = df, x= 'total_bill',y='sex')
```

```python
sns.violinplot(data = df, x= 'total_bill',y='sex',hue='time')
```

# Categorical Plots
### Distributions 

```python
**Strip**-Overlapping 
**Swarm**-Cleaar Plot 
```

```python
df.head(3)
```

```python
sns.catplot(data=df,x='total_bill',y="smoker",kind ='swarm')
```

```python
sns.catplot(data=df,x='total_bill',y="smoker",kind ='strip')
```

```python
sns.catplot(data=df,x='total_bill',y="smoker",kind ='swarm',hue='sex')
```

```python
sns.catplot(data=df,x='tip',y="smoker",kind ='swarm',hue='sex')
```

```python
sns.catplot(data=df,x='tip',y="smoker",kind ='swarm',hue='sex',row ='time',col='day')
```

```python
sns.catplot(data=df,x='tip',y="smoker",kind ='bar',hue='sex',row ='time',col='day')
```
