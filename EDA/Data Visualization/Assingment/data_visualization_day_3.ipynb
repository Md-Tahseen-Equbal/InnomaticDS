###### 2. Write a Python program to display a box plot of the total bill of day and hue = "smoker" and write the intuition about the plot.

```python
### CODE HERE

# Create the box plot using seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=df)

# Add title and labels
plt.title('Total Bill by Day and Smoker')
plt.xlabel('Day')
plt.ylabel('Total Bill')

# Show the plot
plt.show()

# Intuition:
# The box plot shows that the median total bill is higher for smokers than for non-smokers on all days of the week.
# Additionally, the range of total bills is larger for smokers than for non-smokers on all days of the week.
# This suggests that smokers tend to spend more money on meals than non-smokers, regardless of the day of the week.
```

##### 3. Write a Python program to display a violin plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
### code here
# Create the violin plot using seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='total_bill', hue='sex', data=df)

# Add title and labels
plt.title('Total Bill by Day and Sex')
plt.xlabel('Day')
plt.ylabel('Total Bill')

# Show the plot
plt.show()

# Intuition:
# The violin plot shows that the distribution of total bills is similar for males and females on all days of the week.
# However, there is a slightly higher median total bill for males than for females on all days of the week.
# Additionally, the range of total bills is larger for males than for females on all days of the week.
# This suggests that there is more variability in the total bills of males than females, regardless of the day of the week.
```

##### 4. Write a Python program to display a strip plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
### code here
# Create the strip plot using seaborn
plt.figure(figsize=(10, 6))
sns.stripplot(x='day', y='total_bill', hue='sex', data=df)

# Add title and labels
plt.title('Total Bill by Day and Sex')
plt.xlabel('Day')
plt.ylabel('Total Bill')

# Show the plot
plt.show()

# Intuition:
# The strip plot shows the individual data points for the total bill of each customer, by day and sex.
# This allows us to see the distribution of total bills for each group, as well as any outliers.
# We can see that there is a wide range of total bills for both males and females on all days of the week.
# However, there are a few outliers for males, with some customers spending much more than the majority.
```

##### 5. Write a Python program to display heat map of dataframe and write the intuition about the plot.

```python
### code here
# Create a sample dataframe
df = pd.DataFrame({
  'A': [1, 2, 3, 4, 5],
  'B': [4, 5, 6, 7, 8],
  'C': [7, 8, 9, 10, 11],
  'D': [10, 11, 12, 13, 14]
})

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm')

# Add title and labels
plt.title('Heatmap of DataFrame')
plt.xlabel('Columns')
plt.ylabel('Rows')

# Show the plot
plt.show()

# Intuition:
# The heatmap shows the values of the dataframe as a color-coded matrix.
# The color of each cell corresponds to the value of the data point in that cell.
# This allows us to see the overall distribution of the data, as well as any areas where the values are particularly high or low.
# In this case, we can see that the values in the dataframe are generally increasing from left to right and top to bottom.
# There is also a diagonal pattern, with the highest values in the cells along the diagonal.
```

##### 6. Write a Python program to display scatter plot of total bill of tip and write the intuition about the plot.

##### 7. Write a Python program to display density plot of all numerical columns from the dataframe and write the intution about plot.

```python
### code here

# Create density plots for all numerical columns
plt.figure(figsize=(12, 8))
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    sns.kdeplot(df[column], shade=True, label=column)

# Add title and legend
plt.title('Density Plot of All Numerical Columns')
plt.legend()

# Show the plot
plt.show()
'''
Intuition:
The density plot visualizes the distribution of numerical columns in the dataset.
It helps in understanding the distribution and spread of values for each column.
Peaks in the plot represent the most frequent values, while the width of the plot indicates variability.
'''
```

##### 8. Write a Python program to calculate average total bills by days and display top 4 days by plot.

```python
### code here
# Calculate average total bills by days
average_bills_by_day = df.groupby('day')['total_bill'].mean().sort_values(ascending=False).head(4)

# Plot the top 4 days with highest average total bills
plt.figure(figsize=(10, 6))
average_bills_by_day.plot(kind='bar')

# Add title and labels
plt.title('Top 4 Days by Average Total Bill')
plt.xlabel('Day')
plt.ylabel('Average Total Bill')

# Show the plot
plt.show()
```
