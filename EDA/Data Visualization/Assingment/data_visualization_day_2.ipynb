##### import the titanic dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"/content/titanic.csv")
```

```python
df.head(3)
```

##### 1. Write a Python program to display a bar chart of the fare of class.

```python
### code here
# Create a bar chart of the fare of class
sns.barplot(x="class", y="fare", data=df)

# Show the plot
plt.show()
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python
### code here
sns.barplot(x="fare", y="class", data=df)
plt.show()
```

###### 3. Write a Python program to display a bar chart of the fare of class. Use different color for each bar.

```python
### code here
# Set the color palette
sns.set_palette(["red", "green", "blue"])

# Plot the barplot
sns.barplot(x="class", y="fare", data=df, hue="class")
plt.show()
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
### code here
# Create a pie chart of the fare of class
labels = df["class"].value_counts().index.to_list()
sizes = df["class"].value_counts().to_list()
colors = ["red", "green", "blue"]

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%")
plt.title("Pie Chart of the Fare of Class")
plt.show()
```

######5. Write a Python program to display a count plot of all categorical columns.

```python
### code here
# Get all categorical columns
cat_cols = df.select_dtypes(include=["object"]).columns

# Iterate over each categorical column and create a count plot
for col in cat_cols:
  sns.countplot(x=col, data=df)
  plt.show()
```

###### 6. Write a Python program to display a factor plot of ( pclass vs embarked). hist (kind = 'count')

```python
  #### code here
sns.catplot(x="pclass", hue="embarked", kind="count", data=df)
plt.show()
```

###### 7. Write a Python programming to display a factor plot of ( pclass vs sex). hist (kind = 'count')

```python
### code here
sns.catplot(x="pclass", hue="sex", kind="count", data=df)
plt.show()
```

###### 8. Write a Python programming to display a factor plot of ( embarked vs sex). hist (kind = 'count')

```python
### code here
sns.catplot(x="embarked", hue="sex", kind="count", data=df)
plt.show()
```

###### 9. Write a Python program to display distribution plot for all numberical columns.

```python
### code here
# Get all numerical columns
num_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Iterate over each numerical column and create a distribution plot
for col in num_cols:
  sns.distplot(df[col])
  plt.show()
```

###### 10. Write a Python program to display boxplot of age, pclass and hue = gender.

```python
### code here
sns.boxplot(x="age", y="pclass", hue="sex", data=df)
plt.show()
```
