######1. load the iris.csv dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/content/Iris.csv")
```

```python
df.head(3)
```

#### 1. create a bar plot for count the frequency of the three species.

```python
### code here
# Count the frequency of the three species
species_counts = df['Species'].value_counts()
species_counts
```

```python
# Create a bar plot
plt.bar(species_counts.index, species_counts.values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Frequency of the Three Species')
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
### code here
species_counts = df['Species'].value_counts()
plt.pie(species_counts.values, labels=species_counts.index, autopct="%1.1f%%")
plt.title('Percentage of Species')
plt.show()
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here
import seaborn as sns

# Create a pairplot of all numerical variables
sns.pairplot(df)
plt.show()
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here

sns.scatterplot(data=df, x="SepalLengthCm", y="SepalWidthCm", hue="Species")
plt.show()
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
### code here
# Univariate boxplot for SepalLengthCm
sns.boxplot(x=df['SepalLengthCm'])
plt.xlabel('SepalLengthCm')
plt.ylabel('Value')
plt.title('Univariate Boxplot for SepalLengthCm')
plt.show()
```

```python
# Bivariate boxplot for SepalLengthCm and SepalWidthCm
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.xlabel('Species')
plt.ylabel('SepalLengthCm')
plt.title('Bivariate Boxplot for SepalLengthCm and SepalWidthCm')
plt.show()
```

###### 6. Create a Strip plot

```python
### code here
sns.stripplot(x=df['Species'], y=df['SepalLengthCm'])
plt.xlabel('Species')
plt.ylabel('SepalLengthCm')
plt.title('Strip Plot of SepalLengthCm by Species')
plt.show()
```

###### 7. Create a Violin Plot.

```python
### code here
sns.violinplot(x=df['Species'], y=df['SepalLengthCm'])
plt.xlabel('Species')
plt.ylabel('SepalLengthCm')
plt.title('Violin Plot of SepalLengthCm by Species')
plt.show()
```

##### 8. Create a pair chart.

```python
### code here
sns.pairplot(df, hue='Species')
plt.show()
```

##### 9. Create a Heat Map

```python
### code here
# Exclude non-numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

###### 10. Create a distribution plot for all numerical colomns.

```python
### code here
# Create a list of numerical columns
numerical_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Iterate through the numerical columns and create a distribution plot for each
for col in numerical_columns:
  sns.distplot(df[col])
  plt.show()
```
