# Categorical Data

1. Non - Visualization 
  - mode(),Unique(),nunique(),value_count()
2. Visualization
  - Bar Plot
     - How the values / Frequency are Distributed among the categories
  - Count Plot
     - Purpose To count Value 
3. Pie Chart - To Check percentiles / Portions

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = sns.load_dataset("diamonds")
df
```

```python
df['cut'].value_counts()
```

```python
df['cut'].value_counts().plot(kind="bar",
                              color = ["red","green","blue","brown","yellow"],
                             width = 0.8)
plt.ylabel('Counts')
plt.title('Bar Plot of Diamond Cut')
plt.legend()
plt.show()
```

```python
df['color'].value_counts()
```

```python
df['color'].value_counts().plot(kind="bar",
                              color = ["red","green","blue","brown","yellow","Skyblue","pink"],
                             width = 0.8)
plt.ylabel('Counts')
plt.title('Color Plot of Diamond Cut')

plt.show()
```

##  Insight 
1. Most Common Color: The color 'G' has the highest count, making it the most common diamond color.
2. Least Common Color: The color 'J' has the lowest count, indicating it is the least common diamond color.
3. Mid-Range Colors: Colors 'E' and 'F' have similar counts, both falling in the mid-range of frequency.
4. Color Distribution: There is a clear decreasing trend in the count from 'G' to 'J', showing that higher letter grades are less common.

```python
# (2,1,1) 2- It represent a 2 column, 1 It represent row ,1 last it represent a position of plot 
# Multiplication of starting two decide how much plot you have to used 
plt.subplot(2,2,1)
sns.kdeplot(data = df,x='carat')

plt.subplot(2,2,2)
sns.boxplot(data =df,x="carat")

plt.subplot(2,2,3)
sns.countplot(data = df,y='clarity')

plt.subplot(2,2,4)
sns.countplot(data =df,x="color")
```
