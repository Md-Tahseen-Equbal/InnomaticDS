```python
# Merging   --- Go with Youtube 
```

```python
df1
```

```python
df3 = pd.DataFrame({"Name":["Arjun","Akash","Arthi"],
                  "Company":["Accenture","Google","IBM"],
                   "Gender":["Male","Male","Female"],
                   "Married":["Y","N","Y"]}
                  )
df3
```

```python
pd.merge(df1,df3, on = "Company",how = "outer")   # Unioun Data
```

```python
pd.merge(df1,df3, on = "Company",how = "inner")    # Intersection and Give Common Value 
```

```python
pd.merge(df1,df3, on = "Company",how = "left")       # Left + Common 
```

```python
pd.merge(df1,df3, on = "Company",how = "right")     # Right + Common
```
