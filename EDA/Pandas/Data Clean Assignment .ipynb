```python
df['Feature3'].plot(kind= 'box')
```

```python
Q1 = df["Feature3"].quantile(0.25)
Q3 = df["Feature3"].quantile(0.75)
IQR = Q3 - Q1
Lower = max( df["Feature3"].min(), Q1 - 1.5 * IQR)
Upper = min( df["Feature3"].max(), Q3 + 1.5 * IQR)
```

```python
condition = (df["Feature3"] < Lower) |  (df["Feature3"] > Upper)
condition
```

```python
df['Values'].where(~condition,60)
```
