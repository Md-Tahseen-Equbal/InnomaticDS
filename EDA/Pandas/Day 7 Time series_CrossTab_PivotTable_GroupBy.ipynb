# Cross Tab 

### --->Categorical and Categorical

### ---> Frequency, Count , Percentiles

```python
df = pd.read_csv(r"C:\Users\91771\Desktop\Innomatic\EDA\Pandas\Datasets\loandata.csv")
df
```

```python
df.head()
```

```python
df.shape
```

```python
pd.crosstab(df['Gender'],df['Married'])    # Categorical --- categorical Data set Relation
```

```python
pd.crosstab(index=df['Gender'],columns=df['Married'])
```

```python
pd.crosstab(columns=df['Gender'],index=df['Married'])
```

```python
pd.crosstab(columns=df['Gender'],index=df['Married'],normalize=True)*100   # It represent data in percentile 
```

```python
# Cross Tab Relation for --- Education and Loan_Status
```

```python
pd.crosstab(columns=df['Education'],index=df['Loan_Status'])
```

```python
pd.crosstab(columns=df['Education'],
           index=df['Loan_Status'],
           values=df['ApplicantIncome'],
           aggfunc='max')
```

```python
pd.crosstab(columns=df['Education'],
           index=df['Loan_Status'],
           values=df['ApplicantIncome'],
           aggfunc='max',)
```

```python
# Gender self employed and loan_amount  --- Create Cross Tab 
```

```python
pd.crosstab(values=df['LoanAmount'],
            index=df['Self_Employed'],
           columns=df['Gender'],
           aggfunc='mean')
```

```python
pd.crosstab(columns =[df['Education'],df['Gender']],
           index = df['Loan_Status'])
```

# Pivot Table

## ----> Categorical -- Numerical Data

```python
pd.pivot_table(data = df,
               index =df['Education'],
               columns = df['Loan_Status'],
              values = 'ApplicantIncome',
              aggfunc = 'mean')
```

```python
# Difference Pivot Table and Cross Tab
```

```python
pd.pivot_table(data = df,
               columns =df['Education'],
               index = df['Loan_Status'])
```
